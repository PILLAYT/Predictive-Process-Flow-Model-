```python
import simpy, random, pandas as pd

# --------------------------------------------------------------------
# 0)  NETWORK DICTIONARY  --------------------------------------------
# --------------------------------------------------------------------
# from my_network_module import network
# Make sure `network` contains all keys referenced below.
# --------------------------------------------------------------------
# Example:
# network = {'SAW1': {'process_time': 11.38, 'output': 'WIPo_SAW1'}, ... }
# --------------------------------------------------------------------

# --------------------------------------------------------------------
# 1)  USER SETTINGS
# --------------------------------------------------------------------
SIM_TIME      = 7000

INTERARRIVAL  = 0        # arrival interval at every SAW

# --------------------------------------------------------------------
# 2)  GLOBAL COLLECTORS
# --------------------------------------------------------------------
global_item_counter = 0
unit_record   = {}          # UnitID → {'SAW', 'EntryTime', 'ExitTime', 'FinalStorageTime'}
wip_history   = []          # list of {'Time','Node','WIP'}
storage_after_press = []    # FIFO buffer feeding CLs
final_storage = []  
final_wip = {}

def log_wip(env, node, queues):
    wip_history.append({'Time': env.now, 'Node': node, 'WIP': len(queues[node])})

# ─── movement trace helper (prints + logs) ──────────────────────────
movement_log = []     # ordered trace of every movement

def log_move(env, uid, src, dest, action):
    entry = {
        'Time':   env.now,
        'UnitID': uid,
        'From':   src,
        'To':     dest,
        'Action': action        # create / start / finish / transport / batch
    }
    movement_log.append(entry)
    print(f"[{env.now:6.1f}] {uid:>6}  {action:<8}  {src:>22} → {dest}")

# --------------------------------------------------------------------
# 3)  STAGE‑1  (SAWs → WIPo_SAW → WIPi_IH → IH → Press)
# --------------------------------------------------------------------
def saw(env, name, queues):
    proc, outq = network[name]['process_time'], network[name]['output']
    while True:
        yield env.timeout(INTERARRIVAL)
        global global_item_counter
        global_item_counter += 1
        uid = f'U{global_item_counter}'
        unit_record[uid] = {'SAW': name, 'EntryTime': env.now,
                            'ExitTime': None, 'FinalStorageTime': None}
        log_move(env, uid, '', name, 'create')
        yield env.timeout(proc)
        queues[outq].append(uid)
        log_wip(env, outq, queues)
        log_move(env, uid, name, outq, 'finish')

def move(env, wipo, queues):
    dests, tt = network[wipo]['next'], network[wipo]['transport_times']
    if not isinstance(tt, list):
        tt = [tt] * len(dests)
    while True:
        if queues[wipo]:
            uid = queues[wipo].pop(0)
            idx  = min(range(len(dests)), key=lambda i: len(queues[dests[i]]))
            dest = dests[idx]
            yield env.timeout(tt[idx])
            queues[dest].append(uid); log_wip(env, dest, queues)
            log_move(env, uid, wipo, dest, 'transport')
        else:
            yield env.timeout(1)

IH_INPUT = {'IH1': 'WIPi_IH1', 'IH2': 'WIPi_IH2', 'IH3': 'WIPi_IH3'}
def ih(env, name, queues):
    proc, press = network[name]['process_time'], network[name]['output']
    buf = IH_INPUT[name]
    while True:
        if queues[buf]:
            uid = queues[buf].pop(0)
            log_move(env, uid, buf, name, 'start')
            yield env.timeout(proc)
            queues[press].append(uid)
            log_move(env, uid, name, press, 'finish')
        else:
            yield env.timeout(1)

def press(env, name, queues, machine_status):     
    proc = network[name]['process_time']
    while True:
        if machine_status[name] and queues[name]:
            uid = queues[name].pop(0)
            log_move(env, uid, name+'_Q', name, 'start')
            yield env.timeout(proc)
            storage_after_press.append(uid)
            unit_record[uid]['ExitTime'] = env.now
            log_move(env, uid, name, 'storage_after_press', 'finish')
        else:
            yield env.timeout(1)
            
# --------------------------------------------------------------------
# DOWNTIME GENERATOR  (drives OEE for a given machine)
# --------------------------------------------------------------------
def downtime(env, machine, machine_status, OEE,
             mttf=50.0):                        # mean‑time‑to‑failure (min)
    """
    Turns the machine on/off so its long‑run availability matches OEE.
    MTTR = MTTF*(1-OEE)/OEE  (standard steady‑state balance)
    """
    mttr = mttf*(1-OEE)/OEE
    while True:
        machine_status[machine] = True          # up
        yield env.timeout(random.expovariate(1/mttf))
        machine_status[machine] = False         # down
        yield env.timeout(random.expovariate(1/mttr))

# --------------------------------------------------------------------
# 4)  STAGE‑2  (CL → SB → ROD → RID → BR → finished_goods)
# --------------------------------------------------------------------
def cl(env, name, queues):
    proc, wipo = network[name]['process_time'], network[name]['output']
    while True:
        if storage_after_press:
            uid = storage_after_press.pop(0)
            log_move(env, uid, 'storage_after_press', name, 'start')
            yield env.timeout(proc)
            queues[wipo].append(uid); log_wip(env, wipo, queues)
            log_move(env, uid, name, wipo, 'finish')
        else:
            yield env.timeout(1)

def batch_move(env, wipo, queues, batch_size=42):
    dests, tt = network[wipo]['next'], network[wipo]['transport_times']
    if not isinstance(tt, list):
        tt = [tt]*len(dests)
    while True:
        if len(queues[wipo]) >= batch_size:
            batch = [queues[wipo].pop(0) for _ in range(batch_size)]
            idx   = min(range(len(dests)), key=lambda i: len(queues[dests[i]]))
            dest  = dests[idx]
            yield env.timeout(tt[idx])
            queues[dest].extend(batch); log_wip(env, dest, queues)
            for u in batch:
                log_move(env, u, wipo, dest, 'batch')
        else:
            yield env.timeout(1)

def machine(env, name, queues):
    proc, wipo = network[name]['process_time'], network[name]['output']
    in_buf     = MACHINE_INPUT.get(name, name)   # ← use map or default
    while True:
        if queues[in_buf]:
            uid = queues[in_buf].pop(0)
            log_move(env, uid, in_buf, name, 'start')
            yield env.timeout(proc)
            queues[wipo].append(uid); log_wip(env, wipo, queues)
            log_move(env, uid, name, wipo, 'finish')
        else:
            yield env.timeout(1)


def final_batch(env, wipo, queues, batch_size=42):
    tt = network[wipo]['transport_times']
    while True:
        if len(queues[wipo]) >= batch_size:
            batch = [queues[wipo].pop(0) for _ in range(batch_size)]
            yield env.timeout(tt)
            queues['finished_goods'].extend(batch)
            final_storage.extend(batch) 
            for u in batch:
                unit_record[u]['FinalStorageTime'] = env.now      # ← stamp
                log_move(env, u, wipo, 'finished_goods', 'batch')
        else:
            yield env.timeout(1)

# --------------------------------------------------------------------
# 5)  MAIN SIMULATION
# --------------------------------------------------------------------
def run_sim():
    env = simpy.Environment()

    # queue dict for every node
    nodes = list(network.keys()) + ['finished_goods']
    queues = {n: [] for n in nodes}
        
    # status flag for each press (True = up, False = down)
    machine_status = {'P1': True, 'P2': True} 

    # --- stage 1 ---
    for s in ('SAW1','SAW2','SAW3','SAW4'):
        env.process(saw(env, s, queues))
    for w in ('WIPo_SAW1','WIPo_SAW2','WIPo_SAW3','WIPo_SAW4'):
        env.process(move(env, w, queues))
    for ihm in ('IH1','IH2','IH3'):
        env.process(ih(env, ihm, queues))  
        
    # press downtime processes  (already have these)
    env.process(downtime(env, 'P1', machine_status, network['P1']['OEE']))
    env.process(downtime(env, 'P2', machine_status, network['P2']['OEE']))

    # press service processes   ← add machine_status as 4th arg
    env.process(press(env, 'P1', queues, machine_status))
    env.process(press(env, 'P2', queues, machine_status))

    # --- stage 2 ---
    for clm in ('CL1','CL2','CL3','CL4'):
        env.process(cl(env, clm, queues))

    batch_nodes = ['WIPo_CL1','WIPo_CL2','WIPo_CL3','WIPo_CL4',
                   'WIPo_SB1',
                   'WIPo_ROD1','WIPo_ROD2','WIPo_ROD3','WIPo_ROD4','WIPo_ROD5',
                   'WIPo_RID1','WIPo_RID2','WIPo_RID3']
    for b in batch_nodes:
        env.process(batch_move(env, b, queues))

    env.process(machine(env, 'SB1', queues))
    for rod in ('ROD1','ROD2','ROD3','ROD4','ROD5'):
        env.process(machine(env, rod, queues))
    for rid in ('RID1','RID2','RID3'):
        env.process(machine(env, rid, queues))
    for br in ('BR1','BR2','BR3'):
        env.process(machine(env, br, queues))

    for w in ('WIPo_BR1','WIPo_BR2','WIPo_BR3'):
        env.process(final_batch(env, w, queues))   

    # run
    env.run(until=SIM_TIME)

    # ----- capture ending WIP in all WIPo_ / WIPi_ nodes -----
    global final_wip
    final_wip = {node: len(q) 
                 for node, q in queues.items()
                 if node.startswith('WIPo_') or node.startswith('WIPi_')}

    # ------------------- DataFrames ----------------------
    df_units = pd.DataFrame(
        [{'UnitID': uid,
          'SAW': rec['SAW'],
          'ArrivalTime':      rec['EntryTime'],
          'StorageEntryTime': rec['ExitTime'],
          'FinalStorageTime': rec['FinalStorageTime']}
         for uid, rec in unit_record.items()]
    )

    df_wip  = pd.DataFrame(wip_history)
    df_move = pd.DataFrame(movement_log).sort_values('Time').reset_index(drop=True)

    print("\n--- df_units head ---"); print(df_units.head())
    print("\n--- df_wip head ---");   print(df_wip.head())
    print("\n--- df_move head ---");  print(df_move.head())
    
    df_final_wip = pd.DataFrame(
        [{'Node': k, 'EndingWIP': v} for k, v in final_wip.items()]
    ).sort_values('Node').reset_index(drop=True)

    print("\n--- Ending WIP in each WIP buffer ---")
    print(df_final_wip)
    
    return df_units, df_wip, df_move, df_final_wip

# --------------------------------------------------------------------
# 6)  RUN
# --------------------------------------------------------------------

if __name__ == "__main__":
    random.seed(42)
    df_units, df_wip, df_move, df_final_wip = run_sim()
    print("Units that reached finished_goods:", final_storage[:10], "...")
    print("How many?", len(final_storage))
```
