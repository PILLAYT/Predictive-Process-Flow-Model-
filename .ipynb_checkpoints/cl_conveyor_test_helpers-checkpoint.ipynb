# CL‑Conveyor Prototype (helpers‑integrated)
This notebook plugs directly into **plant_sim/helpers.py** so that logging and WIP tracking are identical to your main model.

**Quick start**
1. Make sure `helpers.py` is importable (place the notebook in the same folder or add its parent to `PYTHONPATH`).
2. Run all cells.
3. Adjust parameters (`CONVEYOR_CAP`, `THROTTLE`, etc.) and re‑run to observe different behaviours.



```python

import simpy, importlib, sys, pathlib
import pandas as pd
# Ensure helpers resolves
if not pathlib.Path('helpers.py').exists():
    sys.path.append('..')  # assume notebook is in plant_sim/ directory
from plant_sim import helpers as h
import matplotlib.pyplot as plt
```

```python

# ----------------- Tunable parameters -----------------
SIM_TIME      = 20_000
CL_TIMES      = {'CL1': 8, 'CL2': 8, 'CL3': 8, 'CL4': 8}
SB1_TIME      = 5
COOL_TIME     = 1_440

CONVEYOR_CAP  = 100          # None = unlimited
THROTTLE      = True         # pause CL1 & CL4 at PAUSE_LEVEL
PAUSE_LEVEL   = 100
RESUME_LEVEL  = 50

INITIAL_WIP   = 120          # pre‑cooled units to stress conveyor quickly

final_goods_log = []
```

```python

# ----------------- Model definition -------------------
def cl_conveyor_model(sim_time=SIM_TIME):
    env = simpy.Environment()

    # Queues
    storage_after_press = [(f'U{i}', 0) for i in range(INITIAL_WIP)]
    conveyor = []  # unified WIPo_CL buffer

    # helper to record conveyor length via helpers.log_wip
    def log_conveyor_len():
        h.log_wip(env, 'WIPo_CL', {'WIPo_CL': conveyor})  # minimal dict interface

    # CL machines ---------------------------------------------------
    def cl_proc(name, proc_time):
        while True:
            if THROTTLE and name in ('CL1', 'CL4') and CONVEYOR_CAP is not None:
                if len(conveyor) >= PAUSE_LEVEL:
                    yield env.timeout(1)
                    continue

            # pull cooled unit
            if storage_after_press:
                uid, ready = storage_after_press[0]
                if env.now >= ready:
                    storage_after_press.pop(0)
                else:
                    yield env.timeout(1)
                    continue
            else:
                yield env.timeout(1)
                continue

            h.log_move(env, uid, 'storage_after_press', name, 'start')
            yield env.timeout(proc_time)

            if CONVEYOR_CAP is None or len(conveyor) < CONVEYOR_CAP:
                conveyor.append(uid)
                h.log_move(env, uid, name, 'WIPo_CL', 'finish')
                log_conveyor_len()
            else:
                # conveyor full; put unit back
                storage_after_press.insert(0, (uid, env.now))
                yield env.timeout(1)

    # SB1 -----------------------------------------------------------
    def sb1_proc():
        while True:
            if conveyor:
                uid = conveyor.pop(0)
                h.log_move(env, uid, 'WIPo_CL', 'SB1', 'start')
                log_conveyor_len()
                yield env.timeout(SB1_TIME)
                h.log_move(env, uid, 'SB1', 'finished_goods', 'finish')
                h.unit_record[uid]["FinalGoodsTime"] = env.now
                final_goods_log.append(
                {"UnitId": uid, "FinalGoodsTime": env.now})
            else:
                yield env.timeout(1)

    # Periodic sampler for conveyor length
    def sampler():
        while True:
            log_conveyor_len()
            yield env.timeout(50)

    # Register processes
    for name, pt in CL_TIMES.items():
        env.process(cl_proc(name, pt))
    env.process(sb1_proc())
    env.process(sampler())

    env.run(until=sim_time)

# Run demo
cl_conveyor_model()
print(f"Simulation complete. Last WIPo_CL level = {h.wip_history[-1]['WIP'] if h.wip_history else 0}")
                    
df_final_goods = pd.DataFrame(final_goods_log)
df_final_goods.to_csv('CL_test', index=False)
# (optional tidy-up)
# df_final_goods.sort_values("UnitId", inplace=True, ignore_index=True)

print(df_final_goods.head())
```

```python

# ----------------- Conveyor WIP plot -------------------
times  = [rec['Time'] for rec in h.wip_history if rec['Node'] == 'WIPo_CL']
levels = [rec['WIP']  for rec in h.wip_history if rec['Node'] == 'WIPo_CL']
if times:
    plt.figure(figsize=(10,4))
    plt.step(times, levels, where='post')
    plt.xlabel('Time')
    plt.ylabel('WIPo_CL level')
    plt.title('Conveyor fill level')
    plt.show()
else:
    print("No WIP data recorded – ensure the model ran.")
```
